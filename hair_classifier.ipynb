# View data

```
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```
doc(ImageDataBunch.from_folder)
```

```
data.classes
```

```
data.show_batch(rows=3, figsize=(7,8))
```

```
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

# Train model

```
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```
learn.fit_one_cycle(6,8e-3)
```

```
learn.fit_one_cycle(2,3e-3)
```

```
#Nunca quieres entrenar un modelo donde la train_loss sea 
#mayor que la valid_loss. Significa que el lr es pequeño o las epochs son
#pocas, hay que jugar aumentandolas. Si aun asi no conseguimos una buena
#accuracy, necesitamos mas data.
```

```
learn.save('stage-1')
```

```
learn.unfreeze()
```

```
learn.lr_find()
```

```
#para seleccionar el mejor rango, seleccionamos el que tiene mas bajada
```

```
learn.recorder.plot()
```

```
learn.fit_one_cycle(4, max_lr=slice(2e-6,6e-6))
```

```
learn.save('stage-2')
```

# Interpretation

```
learn.load('stage-2');
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
interp.plot_confusion_matrix()
```
